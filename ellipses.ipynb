```python
import numpy as np #numpy==1.21.4
from sklearn.manifold import MDS #scipy==1.7.2
import cv2
import matplotlib.pyplot as plt
import ot
from ot.gromov import cg,gwggrad,gwloss,init_matrix
from tqdm.notebook import trange,tqdm
import time
from utils import *
```

```python
#parameters
Nx = 10
Ny = 10
N = Nx * Ny
x_len = 50
y_len = 50
```

```python
#Load ellipses
ells = np.loadtxt("./data/ellipses_10_10_50_50_as_flat.txt").reshape((Nx,Ny,x_len,y_len))
```

```python
#create mm-spaces from imgs
ells_flat = ells.reshape(N,x_len,y_len)

posns=[]
heights=[]
Ms=[]

for i in range(N):
    pos,M,height = mm_space_from_img(ells_flat[i])
    posns.append(pos)
    heights.append(height)
    Ms.append(M)
```

```python
#functions needed for GW computation
def f(G):
    return gwloss(constC, hC1, hC2, G)

def df(G):
    return gwggrad(constC, hC1, hC2, G)
```

```python
# Pairwise GW (can also be loaded, see cell below)
gw = np.zeros((N,N))
st = time.time()
for i in trange(N):
    M1 = Ms[i]
    height1 = heights[i]
    for j in range(i+1,N):
        M2 = Ms[j]
        height2 = heights[j]
        
        #Compute Wasserstein coupling
        M_ot = ot.dist(posns[i],posns[j])
        G0 = ot.emd(height1,height2,M_ot/np.max(M_ot))
        
        #Compute GW coupling with initial coupling set to Wasserstein coupling G0
        constC, hC1, hC2 = init_matrix(M1, M2, height1, height2, "square_loss")
        res = cg(height1, height2, 0, 1, f, df, G0, armijo=False, C1=M1, C2=M2, constC=constC,log=True)[0]
        gw[i,j] = gwloss(constC, hC1, hC2, res)

gw = np.sqrt(gw)
gw += gw.T
et = time.time()
print("Calculating all pairwise distances took " + str(np.round((et-st)/60,2)) + " minutes.")

#save
#np.savetxt("./0utputs/gw_ells",gw)
```

```python
#Compute LGW distances for all reference measures
Ps_per_ref = []
LGW_per_ref = []
times_per_ref = []
idx = 0
for k in range(len(refs)):
    supp,M,height = refs[k]
    Ps = []
    Ts = []
    st = time.time()
    for i in range(N):
        #Compute Wasserstein coupling
        M_ot = ot.dist(supp,posns[i])
        G0 = ot.emd(height,heights[i],M_ot/np.max(M_ot))
        
        #Compute GW coupling with initial coupling set to Wasserstein coupling G0
        constC, hC1, hC2 = init_matrix(M, Ms[i], height, heights[i], "square_loss")
        P = cg(height, heights[i], 0, 1, f, df, G0, armijo=False, C1=M, C2=Ms[i], constC=constC)
        Ps.append(P)
        
        #barycentric projection of GW coupling P
        T = (np.divide(P.T,height).T).dot(posns[i])
        Ts.append(T)

    #LGW
    lgw = np.zeros((N,N))
    for i in range(N):
        for j in range(i + 1, N):
            lgw[i, j] = LGW(Ts[i], Ts[j],height,metric="sq_loss")
    lgw += lgw.T
    et = time.time()
    print("Time for LGW with ref space "+ str(k) + ": " + str(np.round((et-st)/60,2)) + " min")
    
    times_per_ref.append(et-st)
    Ps_per_ref.append(Ps)
    LGW_per_ref.append(lgw)
```
